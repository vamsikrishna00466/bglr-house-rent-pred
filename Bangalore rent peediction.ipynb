```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("data/Metro_House_Rent.csv")
df.head()
```

```python
#extract bangalore data only
df1 =df[df['city'].apply(lambda state: state == 'Bangalore')]
```

```python
df2=df1[['area', 'rooms', 'bathroom','furniture', 'rent_amount']]
df2.head()
```

```python
categorical_cols=['area', 'furniture']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
df3=enc.fit(df2)
```

```python
df3
```

```python
df2["furniture"].value_counts()
```

```python
df3 = pd.get_dummies(df2["furniture"], drop_first= True)
df3.head()
```

```python
df4 = pd.get_dummies(df2["area"], drop_first= True)
df4.head()
```

```python
df2.drop(["area", "furniture"], axis = 1, inplace = True)
```

```python
df2
```

```python
data= pd.concat([df2, df3, df4], axis = 1)
```

```python
data
```

```python
data.shape
```

```python
y=data['rent_amount']
x=data.drop(['rent_amount'], axis=1)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=42)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
reggre = GradientBoostingRegressor(random_state=0)
```

```python
reggre.fit(X_train, y_train)
```

```python
reggre.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor(max_depth=2, random_state=0)
```

```python
regr.fit(X_train, y_train)
```

```python
regr.score(X_test, y_test)
```
